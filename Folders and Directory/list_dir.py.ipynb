```python
import os
```

```python
count = 0
def list_dir(path):
    global count
    child = os.listdir(path)
    for item in child:
        item_p = os.path.join(path, item)
        if os.path.isdir(item_p) and count < 100:
            list_dir(item_p)
        else:
            print(item_p)
            count += 1
        
```

```python
list_dir("/users")
```
