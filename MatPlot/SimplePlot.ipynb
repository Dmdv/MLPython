### Simple plot

```python
from matplotlib import pyplot
pyplot.plot([0, 2, 4, 6, 8], "o")
pyplot.ylabel("Values")
pyplot.xlabel("Time")
pyplot.title("Test plot")
pyplot.show()
```

### X and Y coord included

```python
x_value = [0, 1, 2, 4, 54, 65]
y_value = [3, 454, 5, 3, 32, 5]

pyplot.plot(x_value, y_value, "o-")

pyplot.ylabel("Values")
pyplot.xlabel("Time")
pyplot.title("Test plot")
pyplot.show()
```

### Randomly generated Y values

```python
import random
```

```python
x_value = [random.randint(0, 30) for x in range(len(x_value))]
y_value = [random.randint(0, 30) for x in x_value]
```

```python
pyplot.plot(x_value, y_value, "o-")

pyplot.ylabel("Values")
pyplot.xlabel("Time")
pyplot.title("Test plot")
pyplot.show()
```
